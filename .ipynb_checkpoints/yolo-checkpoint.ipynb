```python
import ultralytics
ultralytics.checks()
```

```python
from ultralytics import YOLO
model = YOLO('yolov8s.pt')
```

```python
from pathlib import Path
root = Path('./face_skin_yolo').resolve() # root dataset path

model.train(
    data=str((root/'data.yaml')),
    epochs=100,
    imgsz=640,
    batch=16,
    project='runs',
    name='notebook_train',
    seed=0
)
```
